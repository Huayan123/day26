##### 一个画布多个子图

```python
import numpy as np
x = np.arange(1,100)
fig = plt.figure(figsize=(20,8),dpi=80)
ax1 = fig.add_subplot(2,2,1)
# 绘制子图1
ax1.plot(x,x)
# 绘制图2
ax2 = fig.add_subplot(2,2,2)
ax2.plot(x,x**2)
ax2.grid(color='r',linestyle='--',linewidth=1,alpha=0.3)

ax3 = fig.add_subplot(2,2,3)
ax3.plot(x,np.log(x))


ax4 = fig.add_subplot(2,2,4)
ax4.plot(x,-x)
plt.show()
```
